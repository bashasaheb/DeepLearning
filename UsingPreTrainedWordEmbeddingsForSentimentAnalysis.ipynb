```python
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Sample sentiment analysis data
# Replace this with your actual dataset
text_data = ["I love this movie", "This is the worst film ever", "Great plot and acting", "Terrible acting"]

# Labels: 1 for positive, 0 for negative
labels = [1, 0, 1, 0]

# Step 1: Word Embeddings Setup
# (You can choose another library like Gensim or spaCy)
max_words = 10000  # Maximum number of words to keep
maxlen = 20  # Cuts off reviews after 20 words

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(text_data)
sequences = tokenizer.texts_to_sequences(text_data)

# Step 3: Load Pre-trained Word Embeddings
# (Replace this with your pre-trained word embeddings)
embedding_dim = 100  # Choose the dimension of your pre-trained word embeddings
embedding_matrix = np.random.random((max_words, embedding_dim))

# Step 4: Data Preprocessing
padded_sequences = pad_sequences(sequences, maxlen=maxlen)

# Step 5: Word Embeddings Conversion
# Create a simple neural network for sentiment analysis
model = Sequential()
model.add(Embedding(max_words, embedding_dim, input_length=maxlen, trainable=False))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Step 6: Transfer Learning Application
# Train the neural network with pre-trained embeddings
# Convert labels to numpy array
y_train = np.array(labels)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(np.array(padded_sequences), y_train, test_size=0.25, random_state=42)

# Verify the shapes of the data
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Step 7: Comparison with Random Initialization
# Repeat the experiment with random initialization
model_random_init = Sequential()
model_random_init.add(Embedding(max_words, embedding_dim, input_length=maxlen, trainable=True))  # trainable=True for random initialization
model_random_init.add(Flatten())
model_random_init.add(Dense(1, activation='sigmoid'))
model_random_init.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_random_init.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Step 8: Evaluate the models on the test set
# Evaluate the model with pre-trained embeddings on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"\nModel with Pre-trained Embeddings - Test Accuracy: {test_accuracy:.4f}")

# Evaluate the model with random initialization on the test set
test_loss_random, test_accuracy_random = model_random_init.evaluate(X_test, y_test)
print(f"Model with Random Initialization - Test Accuracy: {test_accuracy_random:.4f}")
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Sample sentiment analysis data
# Replace this with your actual dataset
text_data = ["I love this movie", "This is the worst film ever", "Great plot and acting", "Terrible acting"]

# Labels: 1 for positive, 0 for negative
labels = [1, 0, 1, 0]

# Step 1: Word Embeddings Setup
# (You can choose another library like Gensim or spaCy)
max_words = 10000  # Maximum number of words to keep
maxlen = 20  # Cuts off reviews after 20 words

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(text_data)
sequences = tokenizer.texts_to_sequences(text_data)

# Step 3: Load Pre-trained Word Embeddings
# (Replace this with your pre-trained word embeddings)
embedding_dim = 100  # Choose the dimension of your pre-trained word embeddings
embedding_matrix = np.random.random((max_words, embedding_dim))

# Step 4: Data Preprocessing
padded_sequences = pad_sequences(sequences, maxlen=maxlen)

# Step 5: Word Embeddings Conversion
# Create a simple neural network for sentiment analysis
model = Sequential()
model.add(Embedding(max_words, embedding_dim, input_length=maxlen, trainable=False))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Step 6: Transfer Learning Application
# Train the neural network with pre-trained embeddings
# Convert labels to numpy array
y_train = np.array(labels)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(np.array(padded_sequences), y_train, test_size=0.25, random_state=42)

# Verify the shapes of the data
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Step 7: Comparison with Random Initialization
# Repeat the experiment with random initialization
model_random_init = Sequential()
model_random_init.add(Embedding(max_words, embedding_dim, input_length=maxlen, trainable=True))  # trainable=True for random initialization
model_random_init.add(Flatten())
model_random_init.add(Dense(1, activation='sigmoid'))
model_random_init.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_random_init.fit(X_train, y_train, epochs=5, batch_size=2, validation_data=(X_test, y_test))

# Step 8: Evaluate the models on the test set
# Evaluate the model with pre-trained embeddings on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"\nModel with Pre-trained Embeddings - Test Accuracy: {test_accuracy:.4f}")

# Evaluate the model with random initialization on the test set
test_loss_random, test_accuracy_random = model_random_init.evaluate(X_test, y_test)
print(f"Model with Random Initialization - Test Accuracy: {test_accuracy_random:.4f}")
```
